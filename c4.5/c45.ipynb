```python
from chefboost import Chefboost as chef
```

```python
import pandas as pd

df = pd.read_csv("D:\\AML\\tic-tac-toe.csv")
```

```python
df
```

```python
config = {'algorithm': 'C4.5'}
model = chef.fit(df, config = config, target_label = 'class')
```

```python
df
```

```python
prediction = chef.predict(model, param = ['x','x','x','x','o','o','x','o','o'])
```
